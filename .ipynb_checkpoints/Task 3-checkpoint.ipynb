## Task 3: Selecting features with confidence
### Question 1
For this task we're going to:
1. Apply the technique of Part 2 by generating bootstrap samples and applying lasso selection to each bootstrap sample.
3. Collect count-statistics for each feature.
4. Vary signal strength, sample size, sparsity, $\lambda$.
5. Threshhold the histogram, i.e. identify the features that are selected in more than $X\%$ of the bootstrap samples? Did this identify the true features?

We start by defining a function that runs a parametergrid and our choices of $\lambda$:

```python
# Main imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# "From" imports
from sklearn.model_selection import ParameterGrid, train_test_split
from sklearn.linear_model import LassoCV, Lasso
from sklearn.metrics import confusion_matrix, mean_squared_error as mse
from sklearn.metrics import confusion_matrix as cm


def Lasso_grid(param_grid, alphas, n_iter):
    """Simulate data from parameter grid and run lasso over it

    Parameters
    ----------
    param_grid: A ParameterGrid object containing all combinations of n, p, sparsity etc. 

    """
    fontsize = 15
    for point in param_grid:
        X,y,beta = simulate_data(**point)
        subsample_size = int(np.ceil(0.8*point['n']))
        for alpha in alphas:
            coeff_counts = np.zeros(point['p'])
            for iter in range(n_iter):
                subsample_indices = np.random.choice(point['n'], subsample_size, replace = False)
                X_, y_ = X[subsample_indices,:], y[subsample_indices]    
                lasso = Lasso(alpha = alpha)
                lasso.fit(X_, y_)
                coeff_counts[lasso.coef_ != 0] += 1

            point_ = point
            point_.pop('rng', None)
            plt.figure()
            plt.hist(coeff_counts, bins = point['p'])  
            plt.ylabel("Occurences", fontsize = fontsize)
            plt.xlabel("Counts", fontsize = fontsize) 
            plt.legend([f"Lambda : {alpha}"], fontsize = fontsize)
            plt.xticks(fontsize = fontsize)
            plt.yticks(fontsize = fontsize)
            plt.title(point,fontsize = fontsize)
def simulate_data(n, p, rng, *, sparsity=0.95, SNR=2.0, beta_scale=5.0):
    X = rng.standard_normal(size=(n, p))
    
    q = int(np.ceil((1.0 - sparsity) * p))
    beta = np.zeros((p,), dtype=float)
    beta[:q] = beta_scale * rng.standard_normal(size=(q,))
    
    sigma = np.sqrt(np.sum(np.square(X @ beta)) / (n - 1)) / SNR

    y = X @ beta + sigma * rng.standard_normal(size=(n,))

    idx_col = rng.permutation(p)
    
    return X[:, idx_col], y, beta[idx_col]
```

We start by varying signal strength: $SNR = [0.75, 1.25, 2, 5]$

```python
n_iter = 250 #number of iterations
params = {'p': [1000],                     # Features
          'n': [500],                      # Samples
   'sparsity': [0.85],  
        'SNR': [0.75, 1.25, 2, 5],         # Signal-to-noise 
 'beta_scale': [5],                        # std of beta coeff
        'rng': [np.random.default_rng()]}
#alpha_min =  1.7219
#alpha_1lse = 1.2148
param_grid = ParameterGrid(params)
alphas = [1.2148]
# We choose a parameter permutation
Lasso_grid(param_grid, alphas, n_iter)
plt.show()
```

We see that as we change the SNR from 0.75 to 5 we see that a count-statistic of 0 becomes more prevalent; this is explainably from the fact that as SNR increases the coefficient selection becomes more stable. The underlying linear model becomes clearer and the Lasso coefficients should more or less coalesce around these. Note that at the same time more features are chosen by the lasso at all iterations, following from the same argument.

Next we threshhold the histogram, i.e. identify the features that are selected in more than X\% of the bootstrap samples. Can we then identify the true features? To do this we rewrite our lasso function

```python
def Lasso_grid_treshold(param_grid, alphas, n_iter, threshold):
    """Simulate data from parameter grid and run lasso over it

    Parameters
    ----------
    param_grid : A ParameterGrid object containing all combinations of n, p, sparsity etc. 
    threshold  : The histogram threshold

    """
    fontsize = 15
    for point in param_grid:
        X,y,beta = simulate_data(**point)
        beta_bool = (beta != 0)
        subsample_size = int(np.ceil(0.8*point['n']))
        for alpha in alphas:
            coeff_counts = np.zeros(point['p'])
            for iter in range(n_iter):
                subsample_indices = np.random.choice(point['n'], subsample_size, replace = False)
                X_, y_ = X[subsample_indices,:], y[subsample_indices]    
                lasso = Lasso(alpha = alpha)
                lasso.fit(X_, y_)
                coeff_counts[lasso.coef_ != 0] += 1

            coeff_bool = coeff_counts/n_iter >= threshold
            cfm = confusion_matrix(beta_bool, coeff_bool)
            tn, fp, fn, tp = cfm.ravel()
            print(f"TN: {tn}, FP: {fp}, FN: {fn}, TP: {tp}")
    return
```

```python
n_iter = 250                               #number of iterations
params = {'p': [1000],                     # Features
          'n': [500],                      # Samples
   'sparsity': [0.85],  
        'SNR': [2],                        # Signal-to-noise ratio
 'beta_scale': [5],                        # std of beta coeff
        'rng': [np.random.default_rng()]}
#alpha_min =  1.7219
#alpha_1lse = 1.2148
param_grid = ParameterGrid(params)
alphas = [1.7219]
threshold = 0.8
Lasso_grid_treshold(param_grid, alphas, n_iter, threshold)
```

We note here that $FP+TP = 78$, less than $TP+FN = 151$. The lasso has then constrained the number of nonzero coefficients to less than those that are actually there. This is expected. We may also note that $\frac{FP}{FP+TP}\approx 1/3$, which means our guesses for which coefficients to use aren't too good. We have to take caution here though, because we've only simulated this once - something that is fixed in the next task. By tweaking the parameters (increasing SNR, n for example) one could likely improve the false-positive-rate, while keeping the model parsimonious. 

### Question 2

For this question, we limit the simulation setup to one level of sparsity and signal strength 


1. Use 3-4 threshold values for your histogram (e.g.$\ 80\%, 90\%$ etc) to do the final selection of features for 2-3 values of $\lambda$ - one larger than $\lambda_{1se}$ so that each bootstrap model is sparse and one smaller $\lambda$ so that each bootstrap model is less sparse.
2. Repeat the simulation B times and compute the selection accuracy (sensitivity, specificity) as a function of threshold and 

For this part we again rewrite our function:

```python
def Lasso_grid_treshold_repeat(param_grid, alphas, n_iter, thresholds, B):
    """Simulate data from parameter grid and run lasso over it

    Parameters
    ----------
    param_grid : A ParameterGrid object containing all combinations of n, p, sparsity etc. 
    threshold  : The histogram threshold
    B          : The number of times we simulate again

    """
    fontsize = 15
    for point in param_grid:
        for threshold in thresholds:
            sensitivity = 0
            specificity = 0
            X,y,beta = simulate_data(**point)
            beta_bool = (beta != 0)
            subsample_size = int(np.ceil(0.8*point['n']))
            for alpha in alphas:
                for i in range(B):
                    coeff_counts = np.zeros(point['p'])
                    for iter in range(n_iter):
                        subsample_indices = np.random.choice(point['n'], subsample_size, replace = False)
                        X_, y_ = X[subsample_indices,:], y[subsample_indices]    
                        lasso = Lasso(alpha = alpha)
                        lasso.fit(X_, y_)
                        coeff_counts[lasso.coef_ != 0] += 1
                    coeff_bool = coeff_counts/n_iter >= threshold
                    cfm = confusion_matrix(beta_bool, coeff_bool)
                    tn, fp, fn, tp = cfm.ravel()
                    print(f"TN: {tn}, FP: {fp}, FN: {fn}, TP: {tp}")
                    sensitivity += tp/(tp+fn)
                    specificity += tn/(tn+fp)
                sensitivity = sensitivity/B
                specificity = specificity/B
                print(f"Thrsh: {threshold}, lambda: {alpha} | Sensitivity: {sensitivity}, Specificity: {specificity}")
        
    return
```

```python
n_iter = 250                               #number of iterations
params = {'p': [1000],                     # Features
          'n': [500],                      # Samples
   'sparsity': [0.85],  
        'SNR': [2],                        # Signal-to-noise ratio
 'beta_scale': [5],                        # std of beta coeff
        'rng': [np.random.default_rng()]}
#alpha_min =  1.7219
#alpha_1lse = 1.2148
param_grid = ParameterGrid(params)
alphas = [0.75, 1.2148,1.7219, 3.5]
thresholds = [0.7, 0.8, 0.9]
B = 5
Lasso_grid_treshold_repeat(param_grid, alphas, n_iter, thresholds, B)
```
